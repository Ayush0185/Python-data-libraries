list vs numpy - Time Taken

```python
from time import process_time
```

Time taken by a list

```python
python_list =[i for  i in range(10000)]

start_time = process_time()

python_list = [i + 5 for i in python_list]

end_time = process_time()

print(end_time - start_time)
```

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array += 5

end_time = process_time()

print(end_time - start_time)


```

Initial placeholders in numpy arrays

```python
# create a numpy array of zeroes
x = np.zeros((4,5))
print(x)
```

```python
#creating a numpy arrays of ones

y = np.ones((2,2))
print(y)
```

```python
#arrays of a particular value
z = np.full((5,4),5)
print(z)
```

```python
#create a identity matrix
a = np.eye(4)
print(a)
```

```python
# create a numpy array with random values
b = np.random.random((3,4))
print(b)
```

```python
#random  integers values array with in specific range

c = np.random.randint(10,100,(3,4))
print(c)
```

```python
#array of evenly spaced values -->specifying the number of values required
d = np.linspace(10,30,6)
print(d)
```

```python
# array of evenly spaced values --> specifying the step
d = np.arange(10,30,5)
print(d)
```

```python
# converty a list to a numpy array
list2 = [10,20,30,40,50]

np_array = np.array(list2)
print(np_array)
type(np_array)
```

Array manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
# reshaping a array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```
